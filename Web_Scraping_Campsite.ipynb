```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager  # starting from version 115 we do not need to download driver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import pandas as pd
import numpy as np
```

```python
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service = service)

website = 'https://www.hipcamp.com/en-US/search?goal=leave_it_better&utm_source=adwords&utm_medium=paid_search&utm+campaign=17278985679&utm_content=140382017441&utm_term=camping+app&device=m&gad_source=1&gbraid=0AAAAADl961cldR2KxAqDXQyTcYI2kAOdh&gclid=EAlalQobChMlk_nZrtPKiQMVOxWtBh3ONjHrEAAYASAAEgKTyfD_BwE&mapLayers=nationalParkSystem&mapLayers=nationalForests&mapLayers=publicCampgrounds&searchSource=search-autocomplete&q=Ontario&lat=49.451564&lng=-84.253981&placeId=region.17447&placeName=Ontario%2C+Canada&experience=search'
driver.get(website)
```

```python
# Automate clicking the "Load More" button until all results are shown
while True:
    try:
        # Locate the "Load More" button using XPath or any unique identifier
        show_more_button = driver.find_element(By.XPATH, '//*[@id="__next"]/div[1]/div[4]/div[1]/div[2]/div[2]/button/div')

        # Scroll to the button to ensure it is in view
        ActionChains(driver).move_to_element(show_more_button).perform()

        # Click the button
        show_more_button.click()

        # wait max of 10s until the element (button) is present
        show_more_button = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, '//*[@id="__next"]/div[1]/div[4]/div[1]/div[2]/div[2]/button/div')))
        print("yes")
    except:
        # If the button is not found (indicating all results are loaded), exit the loop
        print("No more 'Load More' buttons found. All results should be loaded.")
        break
```

```python
names = driver.find_elements(By.XPATH, '//*[@id="__next"]//span[contains(@class, "sc-bdvvtL sc-iqseJM  TcHbx")]') #works
prices = driver.find_elements(By.XPATH, '//*[@id="__next"]//span[contains(@class, "sc-bdvvtL sc-iqseJM  eNISBW")]') #works
type_location_provinces = driver.find_elements(By.XPATH, '//*[@id="__next"]//span[contains(@class, "sc-bdvvtL sc-iqseJM  cYXEKZ")]') #works, need to modify in pandas
ratings = driver.find_elements(By.XPATH, '//*[@id="__next"]//span[contains(@class, "sc-bdvvtL sc-iqseJM  iceFHo")]') #works
```

```python
# put the data of type_location_provinces into a list and cleans it
general = [type_location_province.text for type_location_province in type_location_provinces]

unwanted = '', 'Add dates', 'Add guests', '·', 'from', '/ night'
general = [item for item in general if item not in unwanted]
general
```

```python
# creates the location sub-list based on the cleaned list
location_keyword = 'ON'
location = [item for item in general if isinstance(item, str) and location_keyword in item]
location
```

```python
# creates the type_of_campment based on cleaned list
type_of_campment_keyword = '·'
type_of_campment = [item for item in general if isinstance(item, str) and type_of_campment_keyword in item]
site_types = [item.split(' · ')[1] for item in type_of_campment]
site_types
```

```python
# Extract text from each element and store in lists
name = [name.text for name in names]
price = [price.text for price in prices]
type_of_campment = site_types
location = location
rating = [rating.text for rating in ratings]
```

```python
# Clean the price list
price_not_wanted = '', 'Over 500 places'
price = [item for item in price if item not in price_not_wanted]
price
```

```python
# Fine tuning
# Find the maximum length of the lists
max_length = max(len(name), len(rating), len(price), len(type_of_campment), len(location))

# Pad the shorter lists with NaN to match the maximum length
name += [np.nan] * (max_length - len(name))
rating += [np.nan] * (max_length - len(rating))
price += [np.nan] * (max_length - len(price))
type_of_campment += [np.nan] * (max_length - len(type_of_campment))
location += [np.nan] * (max_length - len(location))
```

```python
# Create a pandas DataFrame
df = pd.DataFrame({
    'Name': name,
    'Rating': rating,
    'Price': price,
    'Type': type_of_campment,
    'Location': location,
})
```

```python
# Save the DataFrame to a CSV file
df.to_csv('Ontario_Campsite_Data.csv', index=False)
print("success")
```
